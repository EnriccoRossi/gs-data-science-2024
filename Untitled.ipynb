### Descrição do que será feito:

- Carregar os dados fornecidos das duas bases.
- Combinar as bases para garantir que as informações estejam alinhadas por país e ano.
- Realizar limpeza, como remoção de valores nulos, renomeação de colunas e formatação de números.


```python
import pandas as pd

# Carregar as bases de dados
emissions_data = pd.read_csv("Desktop/FIAP/gs-datascience/Carbon_(CO2)_Emissions_by_Country.csv")
fuels_data = pd.read_csv("Desktop/FIAP/gs-datascience/GCB2022v27_MtCO2_flat.csv")


# Visualizar amostras das duas bases
display(emissions_data.head())
display(fuels_data.head())
```

### Descrição do que será feito:

- Mesclar as duas bases pelo país e ano.
- Ajustar colunas de interesse (e.g., nomes, formatos).
- Explorar os dados para encontrar padrões e realizar análises descritivas.

```python
# Renomear colunas para uniformidade
emissions_data.rename(columns={"Kilotons of Co2": "Emissions_CO2", "Metric Tons Per Capita": "Per_Capita_CO2"}, inplace=True)
fuels_data.rename(columns={"ISO 3166-1 alpha-3": "Country_Code", "Total": "Total_Fuels"}, inplace=True)

country_mapping = {
    "USA": "United States",
}
fuels_data["Country"] = fuels_data["Country"].replace(country_mapping)

# Combinar as bases
combined_data = pd.merge(emissions_data, fuels_data, on=["Country", "Year"], how="inner")

# Exibir dados combinados
display(combined_data.head())
```

### Descrição do que será feito:

- Criar gráficos para obter insights sobre o impacto no meio ambiente.


```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'

# Comparar emissões do Brasil com outras regiões
brasil_data = combined_data[combined_data["Country"] == "Brazil"]
global_avg = combined_data.groupby("Year")["Emissions_CO2"].mean()

plt.figure(figsize=(10, 6))
plt.plot(brasil_data["Year"], brasil_data["Emissions_CO2"], label="Brasil", marker="o")
plt.plot(global_avg.index, global_avg, label="Média Global", linestyle="--")
plt.xlabel("Ano")
plt.ylabel("Emissões de CO2 (Kilotons)")
plt.title("Comparação de Emissões de CO2: Brasil vs. Média Global")
plt.legend()
plt.show()
```

```python
# Filtrar dados
brasil_data = combined_data[combined_data["Country"] == "Brazil"]
america_data = combined_data[combined_data["Region"] == "Americas"]

# Agrupar por ano
america_avg = america_data.groupby("Year")["Emissions_CO2"].mean().reset_index()

# Plotar gráfico
plt.figure(figsize=(10, 6))
plt.plot(brasil_data["Year"], brasil_data["Emissions_CO2"], label="Brasil", marker="o")
plt.plot(america_avg["Year"], america_avg["Emissions_CO2"], label="Média do Continente Americano", marker="s")
plt.title('Brasil x Continente Americano: Emissões de CO₂ ao Longo do Tempo', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Emissões de CO₂ (kilotons)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Filtrar dados
ocde_data = combined_data[combined_data["Country"].isin(["United States", "Canada", "Germany", "France", "Japan"])]  # Exemplo
ocde_avg = ocde_data.groupby("Year")["Per_Capita_CO2"].mean().reset_index()

# Plotar gráfico
plt.figure(figsize=(10, 6))
plt.plot(brasil_data["Year"], brasil_data["Per_Capita_CO2"], label="Brasil", marker="o", color='green')
plt.plot(ocde_avg["Year"], ocde_avg["Per_Capita_CO2"], label="Média da OCDE", marker="s", color='orange')
plt.title('Brasil x Países da OCDE: Emissões de CO₂ Per Capita ao Longo do Tempo', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Emissões de CO₂ Per Capita (toneladas métricas)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Filtrar dados para Brasil e EUA
brasil = combined_data[combined_data["Country"] == "Brazil"]
eua = combined_data[combined_data["Country"] == "United States"]

# Calcular a média anual das emissões de CO₂
brasil_avg = brasil.groupby("Year")["Emissions_CO2"].mean()
eua_avg = eua.groupby("Year")["Emissions_CO2"].mean()

# Configurar o gráfico
plt.figure(figsize=(12, 6))

# Plotar a média de emissões ao longo do tempo
plt.plot(brasil_avg.index, brasil_avg.values, label="Brasil", linestyle='--', color='blue', marker='o')
plt.plot(eua_avg.index, eua_avg.values, label="EUA", linestyle='-', color='red', marker='s')

# Customizar o gráfico
plt.title('Média de Emissões de CO₂ ao Longo do Tempo: Brasil x EUA', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Média de Emissões de CO₂ (kilotons)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

### Descrição do que será feito:

- Utilizar modelos de regressão e classificação para prever emissões futuras.
- Comparar desempenho de pelo menos dois algoritmos de ML.
- Dividir os dados em 70% para treino e 30% para teste.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.impute import SimpleImputer
import numpy as np

# Selecionar features e labels
X = combined_data[["Year", "Total_Fuels", "Oil", "Gas", "Coal"]]
y = combined_data["Emissions_CO2"]

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Resolver valores ausentes (NaN) com imputação da média
imputer = SimpleImputer(strategy="mean")
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# Regressão Linear
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred_reg = regressor.predict(X_test)

# Classificação com Random Forest
# Classificando como 1 (acima da média) ou 0 (abaixo da média)
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train > y_train.mean())  
y_pred_class = classifier.predict(X_test)

# Avaliar os modelos
mse = mean_squared_error(y_test, y_pred_reg)
accuracy = accuracy_score(y_test > y_test.mean(), y_pred_class)

# Exibir os resultados
print(f"Erro quadrático médio (Regressão): {mse}")
print(f"Acurácia (Classificação): {accuracy}")
```
